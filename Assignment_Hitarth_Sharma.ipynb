# 7. Proportion of fully satisfied from CM for each day:

```python
survey.groupby('collection_date').aggregate('count')
```

```python
sg=survey.groupby(['collection_date','CM_satisfaction'])
```

```python
sg1=sg[['response_id']].aggregate('count').unstack()
sg1
```

```python
total_response=sg1.sum(axis=1,level=0)
total_response
```

```python
sg1['Total_Response']=total_response
sg1
```

```python
sg2=sg1['response_id']
sg2
```

```python
total_response=sg2.sum(axis=0,level=0)
total_response
```

# Proportion:

```python
sg2['Proportion']=sg2['Fully Satisfied']/sg2['Total_Response']
sg2
```

# 8. Proportion of fully dissatisfied from MLA for each day:

```python
sm=survey.groupby(['collection_date','MLA_satisfaction'])
sm1=sm[['response_id']].aggregate('count').unstack()
sm1
```

```python
sm2=sm1['response_id']
sm2
```

```python
sm2
```

```python
sm2['Proportion']=sm2['Fully Dissatisfied']/sm2['Total_Response']
```

### Proportion:

```python
sm2
```

### Line Plot:

```python
x=sm2.index.values
y=sm2.Proportion.values
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.title('Proportion of fully dissatisfied with mla',size=20)
plt.xlabel('collection_date')
plt.xticks(rotation=90)
plt.show()
```

# 9.Pivot-table with index as Past_Vote, Column as Vote_Now and cell values as the count of samples.
